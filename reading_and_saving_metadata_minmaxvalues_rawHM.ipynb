Import necessary libraries:

```python
from PIL import Image
from PIL.ExifTags import TAGS
import glob
import os
```

```python
# Directory path where the images are located
directory_path = "./HM Beta TIFF/October 2023/MDCK 17.10.23/B2-1"

# Get a list of all image files in the directory
image_files = glob.glob(os.path.join(directory_path, "*.tiff"))
```

DKD's code for individual images:

```python
image = Image.open(image_files[0])
exifdata = image.getexif()
data=exifdata.get(40092)
data = data.decode('utf-16')
print(len(data))

print(data)
# Close the image to release resources
image.close() 
```

Do the same for each image in the folder, and save the min and max values as a list:

```python
#do this for each image in the directory and save the data as a list
data_list = []
for image_file in image_files:
    image = Image.open(image_file)
    exifdata = image.getexif()
    data=exifdata.get(40092)
    data = data.decode('utf-16')

    data_list.append(data)
    # Close the image to release resources
    image.close()
    
max_values_strings = []
min_values_strings = []


for i in range(len(data_list)):
    max_values_strings.append(data_list[i][34:])
    min_values_strings.append(data_list[i][10:21])
    
max_values_floats = []
min_values_floats = []

for i in range(len(max_values_strings)):
    max_values_strings[i] = float(max_values_strings[i].strip('\x00'))
    min_values_strings[i] = float(min_values_strings[i].strip('\x00'))
    
    max_values_floats.append(max_values_strings[i])
    min_values_floats.append(min_values_strings[i])
    
print(max_values_floats)
print(min_values_floats)
print(len(max_values_floats))
```

Write it to a file to use later:

```python
#save the max and min values as a csv file 
#first column is min values, second column is max values 
import csv

with open('max_min_values_rawHM.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["min_values", "max_values"])
    for i in range(len(max_values_floats)):
        writer.writerow([min_values_floats[i], max_values_floats[i]])

```
